```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
from scipy import io
```

```python
file_path = "./data/IMAGES_RAW.mat"
matdata = io.loadmat(file_path)

images = matdata['IMAGESr'].astype(np.float32)

print(images.shape)
```

```python
img0 = images[:,:,0]
img0.shape
```

```python
for i in range(10):
    plt.figure()
    plt.imshow(images[:,:,i], cmap='gray')
    plt.colorbar()
    plt.show()
```

```python
import cv2

def apply_DoG_filter(gray, ksize, sigma1, sigma2):
    g1 = cv2.GaussianBlur(gray, ksize, sigma1)
    g2 = cv2.GaussianBlur(gray, ksize, sigma2)
    return g1 - g2


for i in range(10):
    img0 = images[:,:,i].astype(np.float32)
    result = apply_DoG_filter(img0, (5,5), 1.3, 2.6)
    
    plt.figure()
    plt.imshow(result, cmap='gray')
    plt.colorbar()
    plt.show()
```
